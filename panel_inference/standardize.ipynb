```python

import pandas as pd

# Step 1: Read your combined panel
df_original = pd.read_excel(r"C:\Users\DELL\OneDrive\Desktop\Project sem 2\Final_Master_Panel_AllColumns.xlsx")

# Step 2: Create a copy for standardization
df_standardized = df_original.copy()

# Step 3: Identify numeric columns automatically (excluding Country, Year, Country Code)
numeric_cols = df_standardized.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Sometimes numbers are read as strings with commas, so clean them
for col in df_standardized.columns:
    if col not in ['Country', 'Year', 'Country Code']:
        # Remove commas and convert to numeric
        df_standardized[col] = pd.to_numeric(df_standardized[col].astype(str).str.replace(',', ''), errors='coerce')

# Step 4: Standardize numeric columns
for col in numeric_cols:
    df_standardized[col] = (df_standardized[col] - df_standardized[col].mean(skipna=True)) / df_standardized[col].std(skipna=True)

# Step 5: Save standardized panel as a new CSV
df_standardized.to_csv(r"C:\Users\DELL\OneDrive\Desktop\Project sem 2\data\Standardized_Panel.csv", index=False)

print("Standardized panel created and saved successfully!")
```
