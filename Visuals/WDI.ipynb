```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from functools import reduce
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import scipy.stats as stats
from scipy.stats import jarque_bera, shapiro

import statsmodels.api as sm
```

```python

def wide_to_long(df,value_name):
    df = df.rename(columns={df.columns[0]: "country"})
    return df.melt(
        id_vars="country",
        var_name="year",
        value_name=value_name
    )

gdp = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\GDP.xlsx", sheet_name="GDP_Cleaned"), value_name="GDP")
unemp = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\Unemployment.xlsx", sheet_name="Unemployment final"), value_name="Unemployment final")
school = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\School enrollment, secondary (% gross).xlsx", sheet_name="SchoolEnrollmentSecondary_Wide"), value_name="SchoolEnrollmentSecondary_Wide")

corruption = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\Control of Corruption.xlsx", sheet_name="Control of Corruption_Wide"), value_name="Control of Corruption_Wide")
gov_eff = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\Government Effectiveness.xlsx", sheet_name="Government Effectiveness_Wide"), value_name="Government Effectiveness_Wide")
pol_stab = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\Political Stability.xlsx", sheet_name="Political Stability_Wide"), value_name="Political Stability_Wide")
rule_law = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\Rule of Law.xlsx", sheet_name="Rule of Law_Wide"), value_name="Rule of Law_Wide")
reg_quality = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\Regulatory Quality.xlsx", sheet_name="Regulatory Quality_Wide"), value_name="Regulatory Quality_Wide")
voice = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\Voice and accountability.xlsx", sheet_name="Voice_and_Accountability_Wide"), value_name="Voice_and_Accountability_Wide")

inequality = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\world inequality database.xlsx", sheet_name="WID_Cleaned"), value_name="WID_Cleaned")
population = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\Total Population.xlsx", sheet_name="Total Population"), value_name="Total Population")
poverty = wide_to_long(pd.read_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\Poverty population(below $3 day).xlsx", sheet_name="PovertyPopulation_Wide"), value_name="PovertyPopulation_Wide")
```

```python
dfs = [
    gdp, unemp, school, corruption, gov_eff, pol_stab,
    rule_law, reg_quality, voice, inequality, population, poverty
]
wdi = reduce(lambda left, right: pd.merge(left, right, on=["country", "year"], how="outer"), dfs)
wdi = wdi.rename(columns={"country": "Country"})

wdi["year"] = wdi["year"].astype(int)
wdi.to_excel("C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\Terrorism_repo\\Data Cleaning\\data\\WDI_Combined.xlsx", index=False)
```
