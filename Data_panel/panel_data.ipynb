```python
import pandas as pd
import os

# ===============================
# 1. FILE PATH
# ===============================
file_path = "C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\data\\School enrollment, secondary (% gross).xlsx"
output_file = "secondary_enrollment_panel.csv"

# ===============================
# 2. LOAD EXCEL SHEETS
# ===============================
# Sheet 0: Enrollment data
df = pd.read_excel(file_path, sheet_name="School enrollment, secondary")

# Sheet 1: Required countries (Column 1)
countries_df = pd.read_excel(file_path, sheet_name="countries", usecols=[0], header=None)
countries_df.columns = ["Country"]

required_countries = (
    countries_df["Country"]
    .dropna()
    .astype(str)
    .str.strip()
    .unique()
    .tolist()
)

# ===============================
# 3. RENAME & CLEAN
# ===============================
# Rename first column to Country
df.rename(columns={df.columns[0]: "Country"}, inplace=True)

# Drop Country Code if present
if "Country Code" in df.columns:
    df.drop(columns=["Country Code"], inplace=True)

# Country name harmonization
country_rename_map = {
    "Côte d’Ivoire": "Cote d'Ivoire",
    "Congo, Rep.": "Congo",
    "Congo, Dem. Rep.": "Congo",
    "Korea, Rep.": "South Korea",
    "Korea, Dem. People’s Rep.": "North Korea",
    "Egypt, Arab Rep.": "Egypt",
    "Iran, Islamic Rep.": "Iran",
    "Venezuela, RB": "Venezuela",
    "Gambia, The": "Gambia",
    "Bahamas, The": "Bahamas",
    "Yemen, Rep.": "Yemen",
    "West Bank and Gaza": "Palestine",
    "Micronesia, Fed. Sts.": "Micronesia",
    "Eswatini (Swaziland)": "Eswatini"
}

df["Country"] = df["Country"].replace(country_rename_map)
df["Country"] = df["Country"].str.strip()

# ===============================
# 4. KEEP ONLY YEAR COLUMNS
# ===============================
year_cols = [c for c in df.columns if str(c).isdigit()]
df = df[["Country"] + year_cols]

# ===============================
# 5. FILTER REQUIRED COUNTRIES
# ===============================
df_filtered = df[df["Country"].isin(required_countries)].copy()

# ===============================
# 6. WIDE → PANEL
# ===============================
panel_df = df_filtered.melt(
    id_vars="Country",
    var_name="Year",
    value_name="Secondary_Enrollment"
)

panel_df["Year"] = panel_df["Year"].astype(int)
panel_df.sort_values(["Country", "Year"], inplace=True)

# ===============================
# 7. SAVE OUTPUT
# ===============================
panel_df.to_csv(output_file, index=False)

# ===============================
# 8. SUCCESS MESSAGE
# ===============================
print("✅ PANEL DATA CREATED SUCCESSFULLY!")
print(f"🌍 Countries used: {panel_df['Country'].nunique()}")
print(f"📄 Total rows: {panel_df.shape[0]}")
print(f"📁 File saved at: {os.path.abspath(output_file)}")
```

```python
import pandas as pd
import os

# ===============================
# 1. FILE PATH
# ===============================
file_path = "C:\\Users\\DELL\\OneDrive\\Desktop\\Project sem 2\\data\\world inequality database.xlsx"   # change if needed
output_file = "world_inequality_panel.csv"

# ===============================
# 2. LOAD EXCEL SHEETS
# ===============================
# Sheet 0: WID data
df = pd.read_excel(file_path, sheet_name="Data")

# Sheet 1: Required countries (Column A)
countries_df = pd.read_excel(file_path, sheet_name="countries", usecols=[0], header=None)
countries_df.columns = ["Country"]

required_countries = (
    countries_df["Country"]
    .dropna()
    .astype(str)
    .str.strip()
    .unique()
    .tolist()
)

# ===============================
# 3. RENAME FIRST COLUMN
# ===============================
df.rename(columns={df.columns[0]: "Country"}, inplace=True)

# ===============================
# 4. DROP NON-COUNTRY IDENTIFIERS
# ===============================
for col in ["Country Code", "ISO", "ISO3", "iso"]:
    if col in df.columns:
        df.drop(columns=[col], inplace=True)

# ===============================
# 5. COUNTRY NAME HARMONIZATION (WID-SPECIFIC)
# ===============================
country_rename_map = {
    "Côte d’Ivoire": "Cote d'Ivoire",
    "Korea": "South Korea",                # WID uses "Korea" generically
    "Iran": "Iran",
    "Viet Nam": "Vietnam",
    "Lao People's Democratic Republic": "Laos",
    "Syrian Arab Republic": "Syria",
    "Bolivia (Plurinational State of)": "Bolivia",
    "Tanzania, United Republic of": "Tanzania",
    "Moldova, Republic of": "Moldova",
    "Congo, Dem. Rep.": "Congo",
    "Congo, Rep.": "Congo",
    "West Bank and Gaza": "Palestine"
}

df["Country"] = df["Country"].replace(country_rename_map)
df["Country"] = df["Country"].str.strip()

# ===============================
# 6. KEEP ONLY YEAR COLUMNS
# ===============================
year_cols = [c for c in df.columns if str(c).isdigit()]
df = df[["Country"] + year_cols]

# ===============================
# 7. FILTER REQUIRED COUNTRIES ONLY
# ===============================
df_filtered = df[df["Country"].isin(required_countries)].copy()

# ===============================
# 8. WIDE → PANEL
# ===============================
panel_df = df_filtered.melt(
    id_vars="Country",
    var_name="Year",
    value_name="Inequality_Measure"
)

panel_df["Year"] = panel_df["Year"].astype(int)
panel_df.sort_values(["Country", "Year"], inplace=True)

# ===============================
# 9. SAVE OUTPUT
# ===============================
panel_df.to_csv(output_file, index=False)

# ===============================
# 10. SUCCESS MESSAGE
# ===============================
print("✅ WORLD INEQUALITY PANEL DATA CREATED SUCCESSFULLY!")
print(f"🌍 Countries retained: {panel_df['Country'].nunique()}")
print(f"📄 Total rows: {panel_df.shape[0]}")
print(f"📁 File saved at: {os.path.abspath(output_file)}")
```

```python
import pandas as pd
import os
import re

# ===============================
# 1. FILE PATHS
# ===============================
input_file = r"C:\Users\DELL\OneDrive\Desktop\Project sem 2\data\Total Population.xlsx"   # CHANGE THIS
population_sheet = "Sheet"      # CHANGE if needed
country_sheet = "countries"                   # Sheet with ~190 countries
output_file = "total_population_panel.csv"

# ===============================
# 2. READ COUNTRY LIST (Sheet 2, Column 1)
# ===============================
country_df = pd.read_excel(input_file, sheet_name=country_sheet, usecols=[0])
country_df.columns = ["Country"]
country_list = country_df["Country"].dropna().unique().tolist()

# ===============================
# 3. READ POPULATION DATA
# ===============================
df = pd.read_excel(input_file, sheet_name=population_sheet)

# Rename first column to Country
df.rename(columns={df.columns[0]: "Country"}, inplace=True)

# ===============================
# 4. FILTER REQUIRED COUNTRIES
# ===============================
df = df[df["Country"].isin(country_list)].copy()

# ===============================
# 5. KEEP ONLY YEAR COLUMNS
# ===============================
year_cols = [
    col for col in df.columns
    if re.fullmatch(r"\d{4}", str(col))
]

df = df[["Country"] + year_cols]

# ===============================
# 6. WIDE → PANEL CONVERSION
# ===============================
panel_df = df.melt(
    id_vars="Country",
    var_name="Year",
    value_name="Total_Population"
)

panel_df["Year"] = panel_df["Year"].astype(int)
panel_df.sort_values(["Country", "Year"], inplace=True)

# ===============================
# 7. SAVE OUTPUT
# ===============================
panel_df.to_csv(output_file, index=False)

# ===============================
# 8. SUCCESS MESSAGE
# ===============================
print("✅ Total Population panel data created successfully!")
print(f"📂 Input file:  {input_file}")
print(f"📁 Output file: {output_file}")
print(f"🌍 Countries used: {panel_df['Country'].nunique()}")
print(f"📊 Total rows: {panel_df.shape[0]}")
```

```python
import pandas as pd
import os

# ===============================
# 1. FILE PATHS
# ===============================
input_file = r"C:\Users\DELL\OneDrive\Desktop\Project sem 2\data\Poverty population(below $3 day).xlsx"     # CHANGE THIS
poverty_sheet = "Poverty population(below $3 day"                   # CHANGE if needed
country_sheet = "countries"                    # Master country list
output_file = "poverty_$3_panel.csv"

# ===============================
# 2. READ MASTER COUNTRY LIST
# ===============================
country_df = pd.read_excel(input_file, sheet_name=country_sheet, usecols=[0])
country_df.columns = ["Country"]
country_list = country_df["Country"].dropna().unique().tolist()

# ===============================
# 3. READ POVERTY DATA
# ===============================
df = pd.read_excel(input_file, sheet_name=poverty_sheet)

# ===============================
# 4. BASIC CLEANING
# ===============================
df.rename(columns={
    "country_name": "Country",
    "reporting_year": "Year",
    "headcount": "Poverty_Headcount"
}, inplace=True)

# Keep only national-level data (important!)
df = df[df["reporting_level"].str.lower() == "national"]

# ===============================
# 5. FILTER REQUIRED COUNTRIES
# ===============================
df = df[df["Country"].isin(country_list)].copy()

# ===============================
# 6. KEEP REQUIRED COLUMNS
# ===============================
panel_df = df[["Country", "Year", "Poverty_Headcount"]]

panel_df["Year"] = panel_df["Year"].astype(int)
panel_df.sort_values(["Country", "Year"], inplace=True)

# ===============================
# 7. SAVE OUTPUT
# ===============================
panel_df.to_csv(output_file, index=False)

# ===============================
# 8. SUCCESS MESSAGE
# ===============================
print("✅ Poverty panel data created successfully!")
print(f"📂 Input file:  {input_file}")
print(f"📁 Output file: {output_file}")
print(f"🌍 Countries used: {panel_df['Country'].nunique()}")
print(f"📊 Total rows: {panel_df.shape[0]}")
```
