```python
import pandas as pd

# Load your poverty data
file_path = r"C:\Users\DELL\OneDrive\Desktop\Project sem 2\Terrorism_repo\Data Cleaning\data\Poverty population(below $3 day).xlsx"

# Step 1: Read poverty data
df = pd.read_excel(file_path, sheet_name='Poverty population(below $3 day')  # replace with your sheet name

# Step 2: Calculate poverty population
df['poverty_population'] = df['headcount'] * df['reporting_pop']

# Step 3: Keep only needed columns
df_clean = df[['region_name', 'region_code', 'country_name', 'country_code', 
               'reporting_year', 'poverty_population']]

# Step 4: Aggregate in case of multiple entries per country-year
df_clean = df_clean.groupby(['country_name', 'reporting_year'])['poverty_population'].sum().reset_index()

# Step 5: Pivot to wide format (countries as rows, years as columns)
df_wide = df_clean.pivot(index='country_name', columns='reporting_year', values='poverty_population')

# Step 6: Read required countries from the "countries" sheet
countries_df = pd.read_excel(file_path, sheet_name='countries')

# Debug: print all column names to see the exact one
print("Columns in 'countries' sheet:", countries_df.columns.tolist())

# Replace 'Country' below with the exact column name you see printed
required_countries = countries_df['COUNTRIES'].tolist()  # <-- change 'Country' if needed

# Step 7: Reindex to include all required countries
df_wide = df_wide.reindex(required_countries)

# Optional: Sort countries alphabetically
df_wide = df_wide.sort_index()

# Step 8: Save to Excel
with pd.ExcelWriter(file_path, mode='a', if_sheet_exists='replace') as writer:
    df_wide.to_excel(writer, sheet_name='PovertyPopulation_Wide')

print("Poverty population data (wide format) saved in sheet 'PovertyPopulation_Wide'. Shape:", df_wide.shape)
```

```python
import pandas as pd

# ===============================
# 1. Load Excel file & sheets
# ===============================
file_path = r"C:\Users\DELL\OneDrive\Desktop\Project sem 2\World Governance indicator.xlsx"   # <-- change this

wgi_df = pd.read_excel(file_path, sheet_name="Voice and Accountability")
country_df = pd.read_excel(file_path, sheet_name="countries")

# ===============================
# 2. Extract list of countries
# ===============================
country_codes = country_df['COUNTRIES'].dropna().unique()

# ===============================
# 3. Filter WGI data by country list
# ===============================
wgi_df = wgi_df[wgi_df['econ_name'].isin(country_codes)]

# ===============================
# 4. Orientation correction
# ===============================
wgi_df['governance_score'] = wgi_df.apply(
    lambda x: x['value'] if x['orientation'] == 1 else 1 - x['value'],
    axis=1
)

# ===============================
# 5. Aggregate to country–year
# ===============================
country_year = (
    wgi_df.groupby(['econ_code', 'econ_name', 'production_year'])['governance_score']
          .mean()
          .reset_index()
)

# ===============================
# 6. Pivot: countries in rows, years in columns
# ===============================
governance_wide = country_year.pivot(
    index=['econ_code', 'econ_name'],
    columns='production_year',
    values='governance_score'
).sort_index(axis=1)

# ===============================
# 7. Save to SAME Excel file
# ===============================
output_path = "C:/Users/DELL/OneDrive/Desktop/Project sem 2/WGI_Governance_Processed.xlsx"

with pd.ExcelWriter(
    output_path,
    engine="openpyxl",
    mode="w"
) as writer:
    governance_wide.to_excel(writer, sheet_name="Voice_and_Accountability_Wide")

print("✅ File saved successfully at:", output_path)


```

```python
import pandas as pd

# Step 1: Load your Excel file
file_path = r"C:\Users\DELL\OneDrive\Desktop\Project sem 2\Terrorism_repo\Data Cleaning\data\world inequality database.xlsx"
wid_sheet = 'Data'          # Sheet containing raw GDP data
countries_sheet = 'countries'  # Sheet containing sorted country list

# Load GDP data (wide format: years as columns)
df_wid = pd.read_excel(file_path, sheet_name=wid_sheet)

# Inspect columns
print("Columns in GDP sheet:", df_wid.columns.tolist())

# Load legitimate countries
df_countries = pd.read_excel(file_path, sheet_name=countries_sheet)
df_wid['Country '] = df_wid['Country '].astype(str).str.strip()
legit_countries = df_countries.iloc[:, 0].astype(str).str.strip().tolist()  # assuming country names are in first column

# Step 2: Filter GDP data for only legitimate countries
df_clean = df_wid[df_wid['Country '].isin(legit_countries)]

# Step 3: Sort by country name
df_clean = df_clean.sort_values(['Country ']).reset_index(drop=True)

# Step 4: Save cleaned data to a new sheet in the same Excel file
with pd.ExcelWriter(file_path, mode='a', if_sheet_exists='replace') as writer:
    df_clean.to_excel(writer, sheet_name='WID_Cleaned', index=False)

print("WID data cleaned and saved in sheet 'WID_Cleaned'. Shape:", df_clean.shape)
```
